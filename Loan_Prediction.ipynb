```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/Data.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.tail()
```

```python
data['Gender']=data['Gender'].map({'Male':1,'Female':0})
```

```python
data['Married']=data['Married'].map({'Yes':1,'No':0})
data['Education']=data['Education'].map({'Graduate':1,'Not Graduate':0})
data['Dependents'].replace('3+',3,inplace=True)
data['Self_Employed']=data['Self_Employed'].map({'Yes':1,'No':0})
data['Property_Area']=data['Property_Area'].map({'Semiurban':1,'Urban':2,'Rural':3})
data['Loan_Status']=data['Loan_Status'].map({'Y':1,'N':0})
```

```python
null_values = data.isnull()
```

```python
null_counts = null_values.sum()
```

```python
print(null_counts)
```

```python
data.mean()
```

```python
data["Gender"] = data["Gender"].fillna(data["Gender"].mode().iloc[0])
```

```python
data["Married"] = data["Married"].fillna(data["Married"].mode().iloc[0])
```

```python
data["Dependents"] = data["Dependents"].fillna(data["Dependents"].mode().iloc[0])
```

```python
data["Self_Employed"] = data["Self_Employed"].fillna(data["Self_Employed"].mode().iloc[0])
```

```python
data["Credit_History"] = data["Credit_History"].fillna(data["Credit_History"].mode().iloc[0])
```

```python
data["LoanAmount"] = data["LoanAmount"].fillna(data["LoanAmount"].mean())
```

```python
data["Loan_Amount_Term"] = data["Loan_Amount_Term"].fillna(data["Loan_Amount_Term"].mean())
```

```python
null_values1 = data.isnull()
null_counts1 = null_values1.sum()
print(null_counts1)
```

```python
data.head()
```

```python
X=data.drop(columns=['Loan_ID','Loan_Status']).values
Y=data['Loan_Status'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
dt.fit(X_train, Y_train)
Y_pred_train = dt.predict(X_train)
```

```python
Y_pred_test = dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
cm_train=confusion_matrix(Y_pred_train,Y_train)
ac_train=accuracy_score(Y_pred_train,Y_train)
f1score_train=f1_score(Y_pred_train,Y_train)
```

```python
print(cm_train)
```

```python
print(ac_train)
```

```python
print(f1score_train)
```

```python
cm_test=confusion_matrix(Y_pred_test,Y_test)
ac_test=accuracy_score(Y_pred_test,Y_test)
f1score_test=f1_score(Y_pred_test,Y_test)
```

```python
print(cm_test)
```

```python
print(ac_test)
```

```python
print(f1score_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion = 'entropy', random_state = 42)
rfc.fit(X_train, Y_train)

# Evaluating on Training set
rfc_pred_train = rfc.predict(X_train)
print('Training Set Evaluation F1-Score=>',f1_score(Y_train,rfc_pred_train))
```

```python
rfc_pred_test = rfc.predict(X_test)
print('Testing Set Evaluation F1-Score=>',f1_score(Y_test,rfc_pred_test))
```
